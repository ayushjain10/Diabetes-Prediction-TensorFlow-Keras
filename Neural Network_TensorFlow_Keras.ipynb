```python
#Building Convoluted Neural Network with Keras as TensorFlow backend using Python implementation.

#Dataset: Pima Indian Diabetes from Kaggle

#It's a classification problem

#Our Aim: Based on the attributes such as Glucose, BloodPressure, SkinThickness, Insulin and more, predict if the person has Diabetes.

#We will build a neural network base don the diabetes.csv file and then predict the accuracy of the model.

#Caution: The main challenge is to select the number of hidden layers for the model.

#If we select more layers, the accuracy may be more for the training data but may not be for the test set.

#That's the case of overfitting the model.
```

```python
import sys

import matplotlib

import seaborn

import sklearn

import pandas

import numpy

#We will build NN using Keras so import Tensorflow & keras
import tensorflow

from tensorflow import keras
```

```python
#Import neccessary libraries

from pandas import scatter_matrix

from matplotlib import pyplot as plt

from sklearn import model_selection

from sklearn.metrics import accuracy_score

from sklearn.metrics import confusion_matrix

#To standardise all the features independently use StandardScaler from sklearn library
from sklearn.preprocessing import StandardScaler

#To create a sequential model in Keras. Sequential model create models layer by layer.
from tensorflow.keras.models import Sequential

#For densely connected NN
from tensorflow.keras.layers import Dense
```

```python
#Read the input file & Analyzing the file

filepath = "/Users/ayushjain/Downloads/Pima Indians Diabetes/diabetes.csv"

dataset = pandas.read_csv(filepath, header = None).values

#Remove the header
dataset = dataset[1:]

#Analyzing the structure of the data

print(dataset.shape)

print(dataset.ndim)

#We have 2D array with 768 rows and 9 columns.
```

```python
#Splitting the data to Train & Test dataset

#Using the train_test_split() to create both training and test data.
#Training data is used to create the model
#Test data is used for validation of the model which was created using training data

#Let's split the full_data in 75-25, where 75% is training data & test_data as 25%
test_data_size = 0.25

#Select all rows and first 8 as the attributes
X = dataset[:,0:8]

#Select all the rows and the 8th as the class label i.e. Diabetes
Y = dataset[:,8]

train_test_list = model_selection.train_test_split(X, Y, test_size = test_data_size, random_state = 10)

X_train_data = train_test_list[0]

X_test_data = train_test_list[1]

Y_train_data = train_test_list[2]

Y_test_data = train_test_list[3]

#We have splitted the data into training and test data.
```

```python
#Creating Neural Network model

#Creating sequential model. Sequential model create models layer by layer.
nn_model = Sequential()

#For the first Dense(), we add a param input_dim, which specifies number of inputs for input layer
#Dense() has an argument units which specifies the input to the current layer.
#Activation function : relu()
#Relu() : This function returns 0 if there is any negative input but in case of positive number, it returns the same number.
#Relu() can be writtern as relu() = max(0, value)

#Let's create first hidden layer
nn_model.add(Dense(units = 7, input_dim = 8, activation = 'relu'))

#Let's now create the output layer.
#The units = 1 as it is a classification problem, we will get the output either 0 or 1, i.e. person has diabetes or not.
#For the output layer, we use sigmoid as the activation function as we need to scale the value.
#sigmoid = 1/(1+e^x)
nn_model.add(Dense(units = 1, activation = 'sigmoid'))

#For the neural network, we have just created a single hidden layer.
#Now we will compile the model. One of the argument is loss
#Since it's a classification problem, we will use cross entropy as the loss. It is most efficient for classification problem.
#We chose optimiser as adam. Adaptive Moment Estimation. Optimiser tunes the parameters to minimise the error.
nn_model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

```

```python
#We have constructed the Neural Network model using Keras.

#Let's evaluate the performance
nn_model_fitted = nn_model.fit(X_train_data, Y_train_data, epochs = 1000, batch_size = X_train_data.shape[0], initial_epoch = 0)

nn_model.evaluate(X_test_data, Y_test_data, verbose=0)
```

```python
#The performance of the model is only 65%. We can further improve the performance of the model.

#Use Dropout Regularization to improve the neural network model

#Dropout Regularization: It is a technique where a neuron at random is dropped and is not used for model creation.
from tensorflow.keras.layers import Dropout

#It can be applied either on Input Layer or Hidden Layers.

#We will create the model using both approaches and select the best model.

#Dropout Regularization on Input layer
#dropout rate : How many neurons to drop at a particular layer.
#Kernel constraint : Threshold weight for the neuron

#We will use Stochastic Gradient Descent Optimisers 
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder

#Creating model

#Creating sequential model. Sequential model create models layer by layer.
nn_model2 = Sequential()

#Input Layer
nn_model2.add(Dense(12, input_dim=8, activation='relu'))
#Adding a dropout 
nn_model2.add(Dropout(0.2))

#Hidden Layer
nn_model2.add(Dense(8, activation='relu'))

#Output layer
nn_model2.add(Dense(1, activation='sigmoid'))

#Compiling the model
nn_model2.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

#We have now the model ready

```

```python
#Let's evaluate the model's performance

nn_model2.fit(X, Y, epochs=150, batch_size=10)

scores = nn_model2.evaluate(X, Y)
print("\n%s: %.2f%%" % (nn_model2.metrics_names[1], scores[1]*100))

#The accuracy of the model increased from 65% to 74.22%. That's a good improvement

#The efficiency of the model was improved using Dropout Regularization.

#Notice that using more detailed insight in the dataset and by changing the parameters such as epochs, 
#number of neural nodes, the performance of the model can be improved much more.

#The main issue is to select the right number of parameters. We also have to keep in mind to not to overfit the model
#using training data. That's the case with some of the models claiming high accuracy.

#We have constructed the Neural Network model using Keras with Droupout Regularization

#In the end, we have created a Neural Network model which can be used to predict whether the person has diabetes or not.

#A Convolution Neural Network was created using Keras with Python implementation using TensorFlow as a backend.
```
